```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
def plot_model_architecture():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Boxes for different stages
    layers = [
        ("Input Features", (1, 5)),
        ("Embedding & Encoding", (3, 5)),
        ("Transformer Encoder (Optional)", (5, 5)),
        ("GRU/LSTM Layer", (7, 5)),
        ("Attention Mechanism", (9, 5)),
        ("Fully Connected Layer", (11, 5)),
        ("Output (Classification)", (13, 5))
    ]

    # Draw boxes
    for label, (x, y) in layers:
        rect = mpatches.FancyBboxPatch((x, y), 2, 1, boxstyle="round,pad=0.1", 
                                       edgecolor="black", facecolor="lightblue")
        ax.add_patch(rect)
        ax.text(x + 1, y + 0.5, label, fontsize=10, ha="center", va="center")

    # Arrows between layers
    for i in range(len(layers) - 1):
        x1, y1 = layers[i][1]
        x2, y2 = layers[i + 1][1]
        ax.annotate("", xy=(x2, y2 + 0.5), xytext=(x1 + 2, y1 + 0.5),
                    arrowprops=dict(arrowstyle="->", lw=1.5))

    # Formatting
    ax.set_xlim(0, 600)
    ax.set_ylim(3, 7)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)
    
    plt.title("Architecture du modèle AdvancedRNN", fontsize=12, fontweight="bold")
    plt.show()

plot_model_architecture()
```
