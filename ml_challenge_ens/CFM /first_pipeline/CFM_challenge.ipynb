```python
from drive.MyDrive.load import load_data
from drive.MyDrive.loader import to_loader
from drive.MyDrive.model import  AdvancedModel
from drive.MyDrive.train import train
from drive.MyDrive.report_end import make_predictions
import torch
```

```python
num_classes = 24
```

```python
X_train,y_train,X_test = load_data(dummy=False)
train_loader,val_loader = to_loader(X_train,y_train,batch_size=64,val_size=0.1)
```

```python
models_GRU_advanced = AdvancedModel(encode_features=[], num_features = 19, num_class=24, embed_features=[0,1,2,3], num_embed_features=[6,3,2,2], d_hidden=256, embedding_dim=8, dropout=0.01)
```

```python
str(models_GRU_advanced)
```

```python
model = models_GRU_advanced
#model = models
optimizer = torch.optim.Adam(model.parameters(), lr=5e-3)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.95)

# Supposons que `dataset` est dÃ©jÃ  un objet PyTorch Dataset avec les donnÃ©es
trained_model = train(
    model,
    train_loader=train_loader,  # Le jeu de donnÃ©es complet
    val_loader=val_loader,
    optimizer=optimizer,
    scheduler=scheduler,
    device = "cuda" if torch.cuda.is_available() else "cpu",
    num_epochs=60,
    beta=0.5,
    alpha=0.3,
)
```

```python
y_pred = make_predictions(model, X_test)
```
